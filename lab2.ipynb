### Median prices of existing single family homes in California: https://www.car.org/marketdata/data/housingdata 

```python
homeprices = pd.read_csv("/Users/conniechou/Library/CloudStorage/OneDrive-Personal/Connie/Homework/GSB544_Bodwin/Labs/medianhomeprices.csv")
homeprices.head()
```

```python
#clean homeprices data, only get SF, SD, LA, Sac columns
ca_homeprices = homeprices[["Mon-Yr", "Sacramento", "San Francisco", "San Diego", "Los Angeles"]]
ca_homeprices_clean = ca_homeprices.copy() #avoiding "making changes on original data" error, make a copy targeted dataset
ca_homeprices_clean
```

```python
#turn date(string) column into datetime, source: ChatGPT
ca_homeprices_clean["Mon-Yr"] = pd.to_datetime(ca_homeprices_clean["Mon-Yr"], format = "%b-%y") #lowercase y=2 digit year, uppercase Y=4 digit year

ca_homeprices_clean
```

```python
#check type
ca_homeprices_clean.dtypes
```

```python
#split mon-yr into separate month, day, year columns, source: stackoverflow
ca_homeprices_clean["year"] = ca_homeprices_clean["Mon-Yr"].dt.year
ca_homeprices_clean["month"] = ca_homeprices_clean["Mon-Yr"].dt.month
ca_homeprices_clean["day"] = ca_homeprices_clean["Mon-Yr"].dt.day
ca_homeprices_clean
```

```python
#check date type
ca_homeprices_clean.dtypes
```

```python
#convert city values from objects to floats

#first remove dollar sign and comma
ca_homeprices_clean["Sacramento"] = ca_homeprices_clean["Sacramento"].str.replace("$", "", regex = False)
ca_homeprices_clean["Sacramento"] = ca_homeprices_clean["Sacramento"].str.replace(",", "", regex = False)
ca_homeprices_clean["Sacramento"] = ca_homeprices_clean["Sacramento"].astype(float)

ca_homeprices_clean["San Francisco"] = ca_homeprices_clean["San Francisco"].str.replace("$", "", regex = False)
ca_homeprices_clean["San Francisco"] = ca_homeprices_clean["San Francisco"].str.replace(",", "", regex = False)
ca_homeprices_clean["San Francisco"] = ca_homeprices_clean["San Francisco"].astype(float)

ca_homeprices_clean["San Diego"] = ca_homeprices_clean["San Diego"].str.replace("$", "", regex = False)
ca_homeprices_clean["San Diego"] = ca_homeprices_clean["San Diego"].str.replace(",", "", regex = False)
ca_homeprices_clean["San Diego"] = ca_homeprices_clean["San Diego"].astype(float)

ca_homeprices_clean["Los Angeles"] = ca_homeprices_clean["Los Angeles"].str.replace("$", "", regex = False)
ca_homeprices_clean["Los Angeles"] = ca_homeprices_clean["Los Angeles"].str.replace(",", "", regex = False)
ca_homeprices_clean["Los Angeles"] = ca_homeprices_clean["Los Angeles"].astype(float)
```

### The two datasets I will be using to merge: avocados_ca_copy_long and ca_homeprices_clean

```python
avocados_ca_copy.head()
```

```python
#make size of avocados from wide to long
avocados_ca_copy_long = avocados_ca_copy.melt(id_vars = ["geography", "type", "average_price", "year", "month", "total_volume"], value_vars = ["small_medium", "large", "xlarge"], var_name = "size", value_name = "number_avocados_sold")
avocados_ca_copy_long
```

```python
ca_homeprices_clean.head()
```

```python
#turn ca_homeprices_clean from wide to long
ca_homeprices_clean_long = ca_homeprices_clean.melt(id_vars = ["year", "month"], value_vars = ["Sacramento", "San Francisco", "San Diego", "Los Angeles"], var_name = "geography", value_name = "median_home_price")
ca_homeprices_clean_long
```

```python
#join avocados_ca, left join on year and month, NEW DATASET FOR GRAPH CALLED AVOCADOS_HOMES
avocados_homes = avocados_ca_copy_long.merge(ca_homeprices_clean_long, on = ["month", "year", "geography"], how = "left")
avocados_homes
```

```python
#calculate amount of money spent on avocados (average_price * total_volume)
avocados_homes["amt_spent_avocados"] = avocados_homes["average_price"] * avocados_homes["total_volume"]
avocados_homes
```

```python
#scatter plot of relationship between average price and median home price
(ggplot(avocados_homes, aes(x = "median_home_price", y = "amt_spent_avocados", fill = "geography"))
 + geom_point(size = 3)
 + xlab("Median Home Price ($)")
 + ylab("Average Sales of Haas Avocados ($)")
 + ggtitle("Amount Sales of Haas Avocados and Median Home Prices between 2015-2020")
 + facet_wrap("year")
 + theme(figure_size = (15,8))
)
```
