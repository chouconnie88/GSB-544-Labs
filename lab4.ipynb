```python
def stateabbr(state):
    """
    Outputs the state abbreviation when given the state full name
    case insensitive and spelling doesn't matter
    
    Parameter
    ---------
    state : str
    A string of the state name
    
    Returns
    -------
    str 
    A string of the state abbreviation
    
    """
    state_name = us.states.lookup(state)
    if state_name is not None:
        return state_name.abbr
    else:
        return "State not found"

    
    
#stateabbr("asdf") unit testing
```

```python
def coffee_shop(url):
    response_shop = requests.get(url)

    soup_shop = BeautifulSoup(response_shop.content, "html.parser")

    # Finding All US Stores
    stores_shop = soup_shop.find_all("li",
                            attrs=
                             {"class": ""})

    # Creating For Loop that Goes Through Each Store
    rows_shop = []

    for store in stores_shop:

        # Adjusting for Content not containing Titles or Relevant Keys
        if store.find("a") != None:
            cells = store.find("a")

            if cells.get('title') != None:
                loc_tag = cells.get('title')

                # Filtering Just to US Locations
                if loc_tag[-2:] == "us" and loc_tag[-6:-4] != "dc":

                  # Find the name of the State
                    store_tag = cells.get_text()

                  # Creating List of Store Tag to Obtain State Name
                    tag_list = store_tag.split(" ")

                    states_list = ['Dakota','Carolina','Hampshire','York','Jersey','Mexico','Island','Virginia']
                    if tag_list[1] not in states_list:
                        state_name = tag_list[0]

                    else:
                        state_name = " ".join(tag_list[0:2])

                    store_locations = re.findall(r'\((.*?)\)', store_tag)[0] # Locations In Between Parentheses
                    store_locations = pd.to_numeric(store_locations) # Converting to Integers

                  # Obtaining Store Name from URL
                    pattern = r'/([^/]+)$'
                    store_name = re.findall(pattern, url_dd)[0]

                  # Append this data.
                    rows_shop.append({
                        "State": state_name,
                        store_name:store_locations
                    })

    # Converting Rows to Dataframe
    shop_df = pd.DataFrame(rows_shop)
    return shop_df
```
